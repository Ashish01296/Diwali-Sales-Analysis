# Exploratory Data Analysis

```python
  #Visualize Gender Count..

#No.of Male and Female
ax = sns.countplot(x=df['Gender'])

for bars in ax.containers:
    ax.bar_label(bars)


```

```python
 #Visualizing Female And Male Amount..

demo = df.groupby(df['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

print(demo)
```

```python
bar = sns.barplot(x='Gender',y='Amount',data=demo)

for bars in bar.containers:
    bar.bar_label(bars)
```

From Above Graphs We Can see Most of The Buyers Are Female And Even Purchasing power of Females are Greather Then Men

# Age

```python
graph = sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in graph.containers:
    graph.bar_label(bars)
```

From the Above Graph We Can See Mostly Female Buyer Age Between 26-35 years

```python
#Total Number Of Orders From Top 10 States..

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sample = sns.barplot(data=sales_state,x='State',y= 'Orders')

for i in sample.containers:
    sample.bar_label(i)
```

```python
#Top 10 Purchase State Amount
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sample = sns.barplot(data=sales_state,x='State',y= 'Amount')

for i in sample.containers:
    sample.bar_label(i)
```

From Above Kerla State  Has No.of Orders Higher Then Haryana But in Term of Purchasing Amount Haryana State has Higher sales/Amount.


# Marital Status

```python
graph = sns.countplot(data=df,x='Marital_Status',hue='Gender')
sns.set(rc={'figure.figsize':(7,5)})

for bars in graph.containers:
    graph.bar_label(bars)
```

# Occupation

```python
#Gender Wise Occupation
graph = sns.countplot(data=df,x='Occupation',hue='Gender')
sns.set(rc={'figure.figsize':(20,5)})

for bars in graph.containers:
    graph.bar_label(bars)
```

From Above Graph We Can See Mostly I.T Sector Women Have Higher Purchase Rate . Now Checking Occupation Based On Martial Status

```python
#Gender Wise Occupation And married Status
graph = sns.countplot(data=df,x='Occupation',hue='Marital_Status')
sns.set(rc={'figure.figsize':(20,5)})

for bars in graph.containers:
    graph.bar_label(bars)
```

```python
#Top 10 No.of.occupation which has higher Purchase Rates.

sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sample = sns.barplot(data=sales_state,x='Occupation',y= 'Amount')

for i in sample.containers:
    sample.bar_label(i)
```

Here I.T Sector Department Have Higher Purchase Rates

# Product Category

```python

sns.set(rc={'figure.figsize':(30,5)})
sample = sns.countplot(data=df,x='Product_Category')

for i in sample.containers:
    sample.bar_label(i)
```

```python
#Top Product_Category by Amount

sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
sample = sns.barplot(data=sales_state,x='Product_Category',y= 'Amount')

for bar in sample.containers:
    sample.bar_label(bar)
```

From Above Graph We can See that Most of the Sold Products are From Food,Clothing And Electronics Category.

```python
#Top 10 Orders/Selling By Product_ID
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_ID',y= 'Orders')

```
